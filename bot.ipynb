```python
import google.generativeai as genai
import re


genai.configure(api_key="--insert api key here--")
model = genai.GenerativeModel("gemini-1.5-flash")

def clean_markdown(response_text):
    """Function to remove Markdown formatting (like * for bold, _ for italics)."""
    cleaned_text = re.sub(r'(\{1,2}|_{1,2})(.?)\1', r'\2', response_text)
    cleaned_text = re.sub(r'[`*_[\]()#]+', '', cleaned_text)
    return cleaned_text

def ask_question(user_input):
    """Send the user input to the Generative AI model and get the response."""
    if not user_input.strip():
        return "Please enter a valid question."

    try:
        # Call the AI model
        response = model.generate_content(user_input)
        ai_response = response.text
        return clean_markdown(ai_response)
    except Exception as e:
        return f"Error: {str(e)}"

def main():
    """Main loop for the chatbot."""
    print("Welcome to the AI Chatbot!")
    print("Type your questions below. Type 'exit' or 'quit' to end the chat.\n")

    while True:
        # Get user input
        user_input = input("You: ").strip()

        # Exit condition
        if user_input.lower() in ["exit", "quit"]:
            print("AI: Goodbye!")
            break

        # Get AI's response and display it
        ai_response = ask_question(user_input)
        print(f"AI: {ai_response}\n")

# Start the chatbot
main()
```
